Top employers in each country

```python
top_employers_by_country = {}

countries = f500['country'].unique()

for country in countries:
    selected_rows = f500[f500['country'] == country].sort_values('employees', ascending=False)
    top_employers_by_country[country] = selected_rows.iloc[0]['company']
    
top_employers_by_country
```

### Top companies by sector based on ROA

```python
f500['roa'] = f500['profits'] / f500['assets']

sectors = f500['sector'].unique()

top_companies_by_sector = {}

for sector in sectors:
    selected_rows = f500[f500['sector'] == sector].sort_values('roa', ascending=False)
    top_companies_by_sector[sector] = selected_rows.iloc[0]['company']
    
top_companies_by_sector
```
